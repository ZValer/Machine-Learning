Columnas

```python
columns = ['age', 'gender', 'grade_level', 'study_hours_per_day',
    'ai_usage_time_minutes', 'ai_tools_used', 'ai_usage_purpose',
    'ai_dependency_score', 'ai_generated_content_percentage',
    'ai_prompts_per_week', 'ai_ethics_score', 'last_exam_score',
    'assignment_scores_avg', 'attendance_percentage',
    'concept_understanding_score', 'study_consistency_index',
    'improvement_rate', 'sleep_hours', 'social_media_hours',
    'tutoring_hours', 'class_participation_score']

cat_columns = [
    'gender',
    'grade_level',
    'ai_tools_used',
    'ai_usage_purpose'
]
```


**Columnas**

| Variable | Descripción | Valor que puede ingresar |
| --- | --- | --- |
| `age` | Edad del estudiante | Número entero |
| `gender` | Género del estudiante | `Female`, `Male`, `Other` |
| `grade_level` | Nivel académico o grado | `1st Year`, `2nd Year`, `3rd Year`, `10th`, `11th`, `12th` |
| `study_hours_per_day` | Horas de estudio diarias | Número decimal o entero de `0-24` |
| `ai_usage_time_minutes` | Tiempo de uso de herramientas de IA | Número entero en minutos |
| `ai_tools_used` | Herramientas de IA utilizadas | Texto o lista: `ChatGPT`, `Gemini`, `Claude`, `Copilot`, `ChatGPT+Gemini`, `Does Not Use AI` |
| `ai_usage_purpose` | Propósito de uso de IA | `Exam Prep`, `Doubt Solving`, `Homework`, `Coding`, `Notes`, `No Specific Purpose` |
| `ai_dependency_score` | Nivel de dependencia de IA | Número de `0-10` |
| `ai_generated_content_percentage` | Porcentaje de contenido generado con IA | Número de `0-100` |
| `ai_prompts_per_week` | Cantidad de prompts enviados semanalmente | Número entero |
| `ai_ethics_score` | Puntaje de uso ético de IA | Número del `0-10` |
| `last_exam_score` | Calificación del último examen | Número entero del `0-100` |
| `assignment_scores_avg` | Promedio de tareas | Número decimal del `0-100` |
| `attendance_percentage` | Porcentaje de asistencia | Número entero o decimal de `0-100` |
| `concept_understanding_score` | Nivel de comprensión de conceptos | Número del `0-10` |
| `study_consistency_index` | Consistencia de estudio | Número decimal o entero del `0-10` |
| `improvement_rate` | Tasa de mejora académica | Número decimal positivo o negativo |
| `sleep_hours` | Horas de sueño | Número decimal o entero de `0-24` |
| `social_media_hours` | Horas en redes sociales | Número decimal o entero de `0-24` |
| `tutoring_hours` | Horas de tutorías | Número decimal o entero|
| `class_participation_score` | Participación en clase | Número entero de `0-10` |


```python
userInput = {}
print("Plese enter the costumer information:")

for column in columns:
  if column in cat_columns:
    userInput[column] = input(f"{column}: ")
  else:
    userInput[column] = float(input(f"{column}: "))
```

Creamos el dataframe con los datos del usuario que son los que se van a predecir con ayuda del modelo

```python
df = pd.DataFrame(userInput, index=[0])
```

Le pasamos los datos al modelo para que nos de una predicción de acuerdo a los datos que le hemos mandado

```python
predicted_score = model.predict(df)[0]

print(f"El alumno tendrá una calificación aproximada de: {predicted_score:.2f}")
```
